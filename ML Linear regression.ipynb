Linear regression

```python
x=df.drop(['MSRP'],axis=1)
x1=pd.DataFrame(x)
x1
```

```python
y1=df[['MSRP']]

```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
model=lr.fit(x1_train,y1_train)
```

```python
y1_pred=model.predict(x1_test)
y1_pred
```

```python
print(y1_test)
```

```python
from sklearn.metrics import r2_score
print('r2_score:',r2_score(y1_test,y1_pred))
```

KNN

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=5)
```

```python
model1=knn.fit(x1_train,y1_train)
```

```python
y_pre=model1.predict(x1_test)
y_pre
```

```python
y1_test
```

```python
from sklearn.metrics import r2_score
print("accuracy :",r2_score(y1_test,y_pre))
```

SVM


```python
from sklearn.svm import SVR
svr=SVR(kernel='linear')
```

```python
model2=svr.fit(x1_train,y1_train)
```

```python
y1_prd=model2.predict(x1_test)
y1_prd
```

```python
y1_test
```

```python
from sklearn.metrics import r2_score
print('svm accuracy :',r2_score(y1_test,y1_prd))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(random_state=42)
```

```python
model3=dt.fit(x1_train,y1_train)
```

```python
y_prd3=model3.predict(x1_test)
y1_test
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_prd3,y1_test))
```
