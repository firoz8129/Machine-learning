```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("regression.csv")
df
```

```python
x1=df.drop('MSRP',axis=1)
```

```python
y1=df['MSRP']
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
model=lr.fit(x1_train,y1_train)
```

```python
y1_pred=model.predict(x1_test)
y1_pred
```

```python
y1_test
```

```python
from sklearn.metrics import r2_score
print('r2 score :',r2_score(y1_test,y1_pred))
```

KNN=k-nearest neighbours

```python
#find best k value
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score
metric_k=[]
```

```python
neighbours=np.arange(1,11)
#k value from 1 to 10
for k in neighbours:
    knn=KNeighborsRegressor(n_neighbors=k)
    model=knn.fit(x1_train,y1_train)
    y1_prd_knn=model.predict(x1_test)
    acc=r2_score(y1_test,y1_prd_knn)
    metric_k.append(acc)
```

```python
metric_k
```

KNN

```python
knn=KNeighborsRegressor(n_neighbors=3)
modelknn=knn.fit(x1_train,y1_train)
y_prdknn=modelknn.predict(x1_test)
accr=r2_score(y1_test,y_prdknn)
accr
```
